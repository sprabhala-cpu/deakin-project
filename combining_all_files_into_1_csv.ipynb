```python
#importing necessary packages for data cleaning and loading
import glob
import pandas as pd
```

```python
files=sorted(glob.glob('/content/drive/MyDrive/data/2007/T*.*'))
```

```python
len(files)
```

```python
dfs = []
for file in files:
    column_name = pd.read_csv(file,index_col=False).columns[0]
    df = pd.read_csv(file,index_col=False,names=["col1","col2"],skiprows=1)
    df["col3"] = column_name
    dfs.append(df)
df = pd.concat(dfs)
```

```python
df.drop(['col2'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df['col3'].unique()
```

```python
df['col3'].unique()#the date column has trailing spaces tab spaces now i am removing it and make it proper and convert into date time format
s=df['col3'].str.replace('t'," ").str.strip(' \t').str.strip('\t ').str.split()
```

```python
dates_str = s.str[1] + ' ' + s.str[0]
from functools import reduce 

def convert_formats_to_datetimes(s1, formats):
    out = [pd.to_datetime(s1, format=x, errors='coerce') for x in formats]
    return reduce(lambda l,r: pd.Series.combine_first(l,r), out)

formats = ['%d/%m/%y %H:%M:%S.%f', '%d/%m/%Y %H:%M:%S.%f',
           '%m/%d/%y %H:%M:%S.%f','%m.%d.%Y %H:%M:%S.%f',
           '%d.%m.%Y %H:%M:%S.%f',
           '%m/%d/%y %M:%S.%f', '%d/%m/%y %M:%S.%f',
           '%m/%d/%Y %M:%S.%f', '%m.%d.%Y %M:%S.%f',
           
           ]

df['date_time'] = convert_formats_to_datetimes(dates_str, formats)
```
